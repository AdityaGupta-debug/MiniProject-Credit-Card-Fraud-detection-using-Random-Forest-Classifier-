```python
fraud = df[df.Class == 1]
not_fraud = df[df.Class == 0]
```

```python
fraud
```

```python
not_fraud.Amount.describe()
```

```python
not_fraud_sample = not_fraud.sample(492)
```

```python
not_fraud_sample.shape
```

```python
new_df = pd.concat([not_fraud_sample, fraud], axis = 0)
```

```python
new_df.head()
```

```python
new_df.shape
```

```python
new_df.Class.value_counts()
```

```python
X = new_df.drop("Class", axis = 1)
```

```python
X
```

```python
Y = new_df['Class']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score

model1 = LogisticRegression()
model1.fit(X_train, Y_train)
Y_pred1 = model1.predict(X_test)
print(r2_score(Y_test, Y_pred1))
```

```python
from sklearn.neighbors import KNeighborsClassifier

model2 = KNeighborsClassifier()
model2.fit(X_train, Y_train)
Y_pred2 = model2.predict(X_test)
print(r2_score(Y_test, Y_pred2))
```

```python
from sklearn.tree import DecisionTreeClassifier

model3 = DecisionTreeClassifier()
model3.fit(X_train, Y_train)
Y_pred3 = model3.predict(X_test)
print(r2_score(Y_test, Y_pred3))
```

```python
from sklearn.svm import SVC
model4 = SVC()
model4.fit(X_train, Y_train)
Y_pred4 = model4.predict(X_test)
print(r2_score(Y_test, Y_pred4))
```

```python
from sklearn.ensemble import RandomForestClassifier

model5 = RandomForestClassifier(max_depth=10, max_samples=0.6)
model5.fit(X_train, Y_train)
Y_pred5 = model5.predict(X_test)
print(r2_score(Y_test, Y_pred5))
```
